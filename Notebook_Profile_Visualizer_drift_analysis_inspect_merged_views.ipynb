## Inspect Merged Data Profile View Metrics

```python
# Create a dataframe for a data dictionary
import pandas as pd
data = {
    "animal": ["cat", "hawk", "snake", "cat"],
    "legs": [4, 2, 0, 4],  # number of legs each animal has!
    "weight": [4.3, 1.8, 1.3, 4.1],
}

df = pd.DataFrame(data)

# Profile the dataframe with whylogs
import whylogs as why

results = why.log(df)
profile = results.profile()
# and create the profile view
prof_view = profile.view()

# Add more data with a second data frame
more_data = {
    "animal": ["snake", "mosquito", "shark", "cat"],
    "legs": [0, 6, 0, 4],
    "weight": [4.3, 0.01, 1.3, 4.1],
}

df = pd.DataFrame(more_data)
# Create a second profile view for it
second_prof_view = why.log(df).profile().view()

# Merge the two profile views
merged_profile_view = prof_view.merge(second_prof_view)

# See the metrics that are generated for a profile view as a pandas dataframe
merged_profile_view.to_pandas()
```
